```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
Data = pandas.read_excel("don't touch me.xls")
```

```python
histogram = Data.hist("Data")
```

```python
hist1 = Data['Data']
hist1 = hist1.values
size = hist1.size/365
m = np.zeros((365,1))
for i in range(m.size):
    for j in range(size-1):
        m[i] += hist1[i + j*365]/size
plt.figure(figsize = (10,10))
plt.plot(m)
```

```python
ar = []
k = 0
p = 0
for i in range(hist1.size):
    if (hist1[i] > 0):
        k += 1
        p = 0
    else:
        p += 1
        if (p == 1):
            ar.append(k)
            k = 0
plt.figure(figsize = (10,10))
plt.plot(ar)
plt.title('Lasting of rainy periods')
```

```python
plt.hist(ar)
plt.title('Histogram of rainy periods')
```

```python
ar = []
k = 0
p = 0
for i in range(hist1.size):
    if (hist1[i] == 0):
        k += 1
        p = 0
    else:
        p += 1
        if (p == 1):
            ar.append(k)
            k = 0
plt.figure(figsize = (10,10))
plt.plot(ar)
plt.title('Intervals between rainy periods')
```

```python
plt.hist(ar)
plt.title('Histogram of intervals')
```

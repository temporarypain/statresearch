```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
Data = pd.read_excel("precipitation.xlsx")
```

```python
year = Data['Year'].values
month = Data['Month'].values
day = Data['Day'].values
t_min = Data['T_min'].values
t_max = Data['T_max'].values
t_mean = Data['T_mean'].values
tot_prec = Data['Total precipitation'].values
numbers = Data['No'].values
```

```python
mean = tot_prec.mean()
size = len(tot_prec)
variance = tot_prec.var()
std = tot_prec.std()
print mean, size, variance, std
```

```python
discrete_prec = np.zeros((size,1))
```

```python
for i in range(size):
    if (tot_prec[i]>0):
        discrete_prec[i] = 1
```

```python
plt.figure(figsize=(20,6))
plt.plot(numbers[:150],discrete_prec[:150])
```

```python
tryal_array = np.zeros((size,1))
```

```python
for i in range(size):
    if (tot_prec[i]>mean):
        tryal_array[i] = 2
    else:
        if (tot_prec[i]<=mean) and (tot_prec[i]>0):
            tryal_array[i] = 1
```

```python
plt.figure(figsize=(20,6))
plt.plot(numbers[:150], tryal_array[:150])
```

```python
tryal_array2 = np.zeros((size,1))
```

```python
for i in range(size):
    if (tot_prec[i]>std):
        tryal_array2[i] = 2
    else:
        if (tot_prec[i]<=std) and (tot_prec[i]>0):
            tryal_array2[i] = 1
```

```python
plt.figure(figsize=(20,6))
plt.plot(numbers[:150], tryal_array2[:150])
```

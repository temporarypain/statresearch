```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
import os
%matplotlib inline
import scipy.stats as stats
import numpy as np
import statsmodels.api as sm # recommended import according to the docs
```

```python
df = pd.read_excel('df2.xlsx')
```

```python
df.head()
```

```python
gamma = df['Gamma']
gamma2 = df['Gamma2']
lambda_ = df['Lambda']
lambda2_ = df['Lambda2']
```

```python
plt.figure(figsize = (7,7))
plt.scatter(gamma,gamma2, color = 'blue',marker = '*', label = 'Gamma, calculated by two methods')
plt.hold()
plt.legend( bbox_to_anchor=(0.75, 1),loc=1, borderaxespad=0.)
plt.xlabel('Gamma')
plt.ylabel('Gamma2')
plt.savefig('gg__.jpg')
```

```python
plt.figure(figsize = (7,7))
plt.scatter(lambda_,lambda2_, color = 'green',marker = '*', label = 'Lambda, calculated by two methods')
plt.hold()
plt.legend( bbox_to_anchor=(0.75, 1),loc=1, borderaxespad=0.)
plt.xlabel('Lambda')
plt.ylabel('Lambda2')
plt.savefig('ll__.jpg')
```

```python
plt.figure(figsize = (7,7))
plt.scatter(gamma,lambda_, color = 'blue',marker = '*', label = 'OLS-method ')
plt.hold()
plt.scatter(gamma2,lambda2_, color = 'green', marker = 'o', label = 'Quantile method ')
plt.legend( bbox_to_anchor=(0.75, 1),loc=1, borderaxespad=0.)
plt.xlabel('Gamma')
plt.ylabel('Lambda')
plt.savefig('gl_.jpg')
```
